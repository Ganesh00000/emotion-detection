```python
import cv2
import tensorflow as tf
from tensorflow import keras
```

```python
file="emotion.json"
```

```python
f=open("emotion.json","r")
loaded_model=f.read()
f.close
```

```python
import tensorflow_hub as hub
```

```python
from tensorflow_hub import KerasLayer
```

```python
import h5py
```

```python
model=keras.models.model_from_json(loaded_model,custom_objects={'KerasLayer': hub.KerasLayer})
model.load_weights("emotion.h5")
```

```python
import numpy as np
```

```python
facecascade=cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_default.xml")
```

```python
cap=cv2.VideoCapture(0)
while True:
    check,img=cap.read()
    
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face=facecascade.detectMultiScale(gray,1.2,5)
    for x,y,h,k in face:
        cv2.rectangle(img,(x,y),(x+h,y+k),(255,0,0),1)
    image=cv2.resize(img,(224,224))
    image=np.array(image)
    image=image/255
    image=image.reshape(1,224,224,3)
    y_predict=model.predict(image)
    print(y_predict)
    if np.argmax(y_predict)==0:
        result="angry"
    elif np.argmax(y_predict)==1:
        result="disgusted"
    elif np.argmax(y_predict)==2:
        result="fearful"
    elif np.argmax(y_predict)==3:
        result="happy"
    elif np.argmax(y_predict)==4:
        result="neutral"
    elif np.argmax(y_predict)==5:
        result="sad"
    else:
        result="surprised"
    cv2.putText(img,result,(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),1,cv2.LINE_AA)
    cv2.imshow("emotion",img)
    if cv2.waitKey(1) & 0xFF == ord("f"):
        break
cap.release()
cv2.destroyAllWindows()
```
